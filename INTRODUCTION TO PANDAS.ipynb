## UPDATING ROWS AND COLUMNS 
- Renaming rows and columns

### RENAMING COLUMNS

```python
df.rename(columns={'Country': 'countries',
                    'Salary': 'Salaries'}, inplace=True)
df
```

```python
df.columns = ['Country', 'Age', 'Salary', 'Purchased']
df
```

### RENAMING ROWS
- Using replace and loc

```python
#using replace function
df['Country'] = df['Country'].replace({'France':'Italy'})
df
```

```python
df['Country'] = df['Country'].replace({'Spain':'Austria'})
df
```

```python
#using loc function
df.loc[0, 'Country'] = 'France'
df
```

```python
df.loc[5, 'Country'] = 'France'
df
```

```python
df.at[7, 'Country'] = 'France'
df
```

```python
df.at[9, 'Country'] = 'France'
df
```

### CONVERTING COLUMNS AND ROWS TO LOWER OR UPPER CASE

```python
df['Country'] = df['Country'].str.upper()
df
```

```python
#Converting columns to upper/lower case
df['Country'] = df['Country'].str.lower()
df
```
